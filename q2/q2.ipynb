```python
import matplotlib.pyplot as plt
import csv
import random as r

def main():
    f1=open('randomdice.csv','w', newline='', encoding='cp949')
    wr=csv.writer(f1)
    for i in range(0,100):
        num=r.randrange(1,7)
        wr.writerow(['100', num])
    for i in range(0,1000):
        num=r.randrange(1,7)
        wr.writerow(['1000', num])
    for i in range(0,10000):
        num=r.randrange(1,7)
        wr.writerow(['10000', num])
    for i in range(0,100000):
        num=r.randrange(1,7)
        wr.writerow(['100000', num])
    
    f1.close()
    
    f2=open('randomdice.csv','r',encoding='cp949')
    data=csv.reader(f2)
    list1=[]
    list2=[]
    list3=[]
    list4=[]
    
    for row in data:
        if int(row[0])==100:
            list1.append(int(row[1]))
        if int(row[0])==1000:
            list2.append(int(row[1]))
        if int(row[0])==10000:
            list3.append(int(row[1]))
        if int(row[0])==100000:
            list4.append(int(row[1]))
    
    f2.close()
    plt.title('Dice Roll Result, 100 times')
    plt.xlabel('the spot on a die')
    plt.ylabel('numbers of event')
    plt.hist(list1)
    plt.show()
    plt.title('Dice Roll Result, 1,000 times')
    plt.xlabel('the spot on a die')
    plt.ylabel('numbers of event')
    plt.hist(list2)
    plt.show()
    plt.title('Dice Roll Result, 10,000 times')
    plt.xlabel('the spot on a die')
    plt.ylabel('numbers of event')
    plt.hist(list3)
    plt.show()
    plt.title('Dice Roll Result, 100,000 times')
    plt.xlabel('the spot on a die')
    plt.ylabel('numbers of event')
    plt.hist(list4)
    plt.show()
    
    print('100번 시행한 결과에서는 주사위 눈 간의 빈도수 차이가 많이 났지만, 시행 횟수가 많아질수록 각 눈 간의 빈도수 차이가 줄어들어 거의 같은 빈도수를 보인다. 이를 통해 시행 횟수가 많을수록 각 눈이 나올 확률은 동일해진다는 결론을 낼 수 있다.')
    
if __name__=="__main__":
    main()
```
