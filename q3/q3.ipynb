```python
import csv
import matplotlib.pyplot as plt

def main():
    male=[]
    female=[]
    year=[2022,2018,2014,2010]
    f=open('2022_Jeju_population.csv','r',encoding='utf-8')
    data=csv.reader(f,delimiter=',')
    header=next(data)
    temp=0
    for row in data:
        if temp==0:
            male.append(-int(row[3]))
            female.append(int(row[4]))
            temp=1
    f.close()
    
    f=open('2018_Jeju_population.csv','r',encoding='utf-8')
    data=csv.reader(f,delimiter=',')
    header=next(data)
    temp=0
    for row in data:
        if temp==0:
            male.append(-int(row[3]))
            female.append(int(row[4]))
            temp=1
    f.close()
        
    f=open('2014_Jeju_population.csv','r',encoding='utf-8')
    data=csv.reader(f,delimiter=',')
    header=next(data)
    temp=0
    for row in data:
        if temp==0:
            male.append(-int(row[3]))
            female.append(int(row[4]))
            temp=1
    f.close()
    
    f=open('2010_Jeju_population.csv','r',encoding='utf-8')
    data=csv.reader(f,delimiter=',')
    header=next(data)
    temp=0
    for row in data:
        if temp==0:
            male.append(-int(row[3]))
            female.append(int(row[4]))
            temp=1
    f.close()
    
    plt.title('Jeju Population Statistics')
    plt.barh(year,male,label='Male',color='purple')
    plt.barh(year,female,label='Female',color='yellow')
    plt.ylabel('Year')
    plt.xlabel('population')
    plt.legend()
    plt.show()
    print('그래프를 보면 시간이 지남에 따라 인구가 증가하는 추세이다. 여기서 2010년의 남녀인구는 모두 약 28만으로 비슷한 수치를 보이고 있으며 2014년은 약 30만, 2018년은 약 33만, 2022년은 약 34만으로 모든 년도에서 남녀가 비슷한 수치를 보이고 있다. 따라서 제주도에는 여성의 비율이 더 높다는 속설은 틀렸다는 결론을 낼 수 있다.')
    
if __name__=="__main__":
    main()
```
