```python
import csv
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

def main():
    ride={0:0}
    quit={0:0}
    both={0:0}
    max_ride_station=[]
    max_ride_num=[]
    max_quit_station=[]
    max_quit_num=[]
    max_both_station=[]
    max_both_num=[]
    f=open('subway_user_per_time.csv','r',encoding='utf-8')
    data = csv.reader(f)
    header = next(data)
    header = next(data)
    for row in data:
        ride[row[3]]=int(row[10])+int(row[12])
        quit[row[3]]=int(row[11])+int(row[13])
        both[row[3]]=int(row[11])+int(row[13])+int(row[10])+int(row[12])
        
    for i in range(30):
        maxkey=max(ride, key=ride.get)
        max_ride_station.append(maxkey)
        max_ride_num.append(ride[maxkey])
        del ride[maxkey]
        
        maxkey=max(quit, key=quit.get)
        max_quit_station.append(maxkey)
        max_quit_num.append(quit[maxkey])
        del quit[maxkey]
        
        maxkey=max(both, key=both.get)
        max_both_station.append(maxkey)
        max_both_num.append(both[maxkey])
        del both[maxkey]
        
        
    font_path = "C:\Windows\Fonts\HMKMG.TTF"
    font = font_manager.FontProperties(fname=font_path).get_name()
    rc('font', family=font)
    

    plt.title('Maximum Riding Station, 30 stations-1')
    plt.bar(max_ride_station[0:5], max_ride_num[0:5])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Riding Station, 30 stations-2')
    plt.bar(max_ride_station[5:10], max_ride_num[5:10])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Riding Station, 30 stations-3')
    plt.bar(max_ride_station[10:15], max_ride_num[10:15])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Riding Station, 30 stations-4')
    plt.bar(max_ride_station[15:20], max_ride_num[15:20])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Riding Station, 30 stations-5')
    plt.bar(max_ride_station[20:25], max_ride_num[20:25])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Riding Station, 30 stations-6')
    plt.bar(max_ride_station[25:30], max_ride_num[25:30])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Quit Station, 30 stations-1')
    plt.bar(max_quit_station[0:5], max_quit_num[0:5])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Quit Station, 30 stations-2')
    plt.bar(max_quit_station[5:10], max_quit_num[5:10])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Quit Station, 30 stations-3')
    plt.bar(max_quit_station[10:15], max_quit_num[10:15])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Quit Station, 30 stations-4')
    plt.bar(max_quit_station[15:20], max_quit_num[15:20])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Quit Station, 30 stations-5')
    plt.bar(max_quit_station[20:25], max_quit_num[20:25])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Quit Station, 30 stations-6')
    plt.bar(max_quit_station[25:30], max_quit_num[25:30])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Both Station, 30 stations-1')
    plt.bar(max_both_station[0:5], max_both_num[0:5])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Both Station, 30 stations-2')
    plt.bar(max_both_station[5:10], max_both_num[5:10])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Both Station, 30 stations-3')
    plt.bar(max_both_station[10:15], max_both_num[10:15])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Both Station, 30 stations-4')
    plt.bar(max_both_station[15:20], max_both_num[15:20])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Both Station, 30 stations-5')
    plt.bar(max_both_station[20:25], max_both_num[20:25])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()
    
    plt.title('Maximum Both Station, 30 stations-6')
    plt.bar(max_both_station[25:30], max_both_num[25:30])
    plt.xlabel('station name')
    plt.ylabel('guest number')
    plt.legend(["Population"])
    plt.show()

    print('         ,  35    .      ,  55    . ,                68   .              .')
    
if __name__=="__main__":
    main()
```
